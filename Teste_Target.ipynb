````python
#Questão 3 - Target
import numpy as np

faturamento_diario = np.random.choice([0] + list(range(100, 601)), size=365)


dias_contaveis = faturamento_diario[faturamento_diario > 0]

menor = dias_contaveis.min() if dias_contaveis.size > 0 else 0
maior = dias_contaveis.max() if dias_contaveis.size > 0 else 0
media = dias_contaveis.mean() if dias_contaveis.size > 0 else 0

print("Menor Faturamento:", menor)
print("Maior Faturamento:", maior)
print("Média Anual de Faturamento:", media)
````
